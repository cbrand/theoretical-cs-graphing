```python
%matplotlib inline
import matplotlib.pyplot as plt

from networkx import Graph, draw

from scnet.grapher.reader import Reader
from scnet.grapher.graph.drawer import GraphDrawer
```

```python
from io import StringIO
graph = Reader(StringIO("""
# 9 Knotendefinitionen, A bis I
knoten A
knoten B 
knoten C
knoten D
knoten E
knoten F
knoten G
knoten H
knoten I

# Kantendefinitionen 
kante A B 2
kante A F 9
kante A G 15
kante B C 4
kante B G 6
kante C D 2
kante C I 15
kante D E 1
kante D I 1
kante E F 6
kante E H 3
kante F H 11
kante G H 15
kante G I 2
kante H I 4
""")).parse()
GraphDrawer().show()
```

```python
from scnet.grapher.algorithms.minimum_spanning_tree import MinimumSpanningTree

new_graph = MinimumSpanningTree(graph).run()
GraphDrawer(node_store=new_graph.node_store, edge_store=new_graph.edge_store).show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

from networkx import Graph, draw

from scnet.grapher.reader import Reader
from scnet.grapher.graph.drawer import GraphDrawer
```

```python
from io import StringIO
graph = Reader(StringIO("""
# 9 Knotendefinitionen, A bis I
knoten A
knoten B 
knoten C
knoten D
knoten E
knoten F
knoten G
knoten H
knoten I

# Kantendefinitionen 
kante A B 2
kante A F 9
kante A G 15
kante B C 4
kante B G 6
kante C D 2
kante C I 15
kante D E 1
kante D I 1
kante E F 6
kante E H 3
kante F H 11
kante G H 15
kante G I 2
kante H I 4
""")).parse()
GraphDrawer().show()
```

```python
from scnet.grapher.algorithms.minimum_spanning_tree import MinimumSpanningTree

new_graph = MinimumSpanningTree(graph).run()
GraphDrawer.from_graph(new_graph).show()
```

```python
from scnet.grapher.generator import generate_graph
graph = generate_graph(
    ("A","B","C","D","E","F","G","H","I",),
    (
        ("A", "B", 2),
        ("A", "I", 3),
        ("I", "C", 4),
        ("I", "G", 5),
        ("C", "D", 6),
        ("C", "E", 7),
        ("C", "F", 8),
        ("E", "H", 10),
        ("H", "G", 11),
        ("F", "G", 12),
        ('A', 'D', 1),
    )
)
GraphDrawer.from_graph(graph).show()
```

```python
from scnet.grapher.algorithms.minimum_spanning_tree import MinimumSpanningTree

new_graph = MinimumSpanningTree(graph).run()
GraphDrawer.from_graph(new_graph).show()
```

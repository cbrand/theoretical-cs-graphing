```python
from scnet.grapher.algorithms.depth_first_algorithm import DepthFirstAlgorithm
%matplotlib inline
import matplotlib.pyplot as plt

from networkx import Graph, draw

from scnet.grapher.reader import Reader
from scnet.grapher.graph.drawer import GraphDrawer
from scnet.grapher.generator import generate_graph
```

```python
graph = generate_graph(
    ("A","B","C","D","E","F","G","H","I",),
    (
        ("A", "B"),
        ("A", "I"),
        ("I", "C"),
        ("I", "G"),
        ("C", "D"),
        ("C", "E"),
        ("C", "F"),
        ("E", "H"),
        ("H", "G"),
        ("F", "G"),
    )
)
GraphDrawer(node_store=graph.node_store, edge_store=graph.edge_store).show()
```

```python
depth_first_search = DepthFirstAlgorithm(graph)
depth_first_search.traverse()
print([node.name for node in depth_first_search.traverse_tree])
```

```python
from scnet.grapher.algorithms.depth_first_algorithm import DepthFirstAlgorithm
%matplotlib inline
import matplotlib.pyplot as plt

from networkx import Graph, draw

from scnet.grapher.reader import Reader
from scnet.grapher.graph.drawer import GraphDrawer
from scnet.grapher.generator import generate_graph
from scnet.grapher.detection.cycle_detection import CycleDetection
```

```python
graph1 = generate_graph(
            ("A", "B", "C", "D"),
            (
                ("A", "B"),
                ("B", "C"),
                ("C", "D"),
                ("D", "A"),
            )
)
GraphDrawer(node_store=graph1.node_store, edge_store=graph1.edge_store).show()
```

```python
cycle_detection = CycleDetection(graph1, start_at=graph1.node_store.get('A'))
print("Has Cycle: %s" % cycle_detection.detect())
print("Cycle: %s" % ([node.name for node in cycle_detection.cycle]))
```
